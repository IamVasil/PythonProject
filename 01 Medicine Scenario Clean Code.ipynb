### Libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt 
pd.options.mode.chained_assignment = None 
from datetime import datetime
```

### 1. In the Drive folder you have a dataset 'Diabetes_data.csv', read the dataset into Pandas.

```python
df = pd.read_csv("Diabetes_data.csv")
df
```

### 2. The data is slopily gathered and you need to rename, correct, and handle missing values. Check to see if any of the columns have null/missing values and determine the best strategy for cleaning up the data (including not doing anything).

```python
##Renameing columns
df = df.rename(columns={'Kolku mu e tromesechen (hemoglobin A1C) she4er' : 'Kolku mu e tromesechen (hemoglobin A1C) sejker'}) 
df.rename(columns={'Kolku mu e tromesechen (hemoglobin A1C) she4er' : 'Kolku mu e tromesechen (hemoglobin A1C) sejker'},inplace=True) 
df = df.rename(columns={"Unnamed: 0":"Patient ID"}) 
df.rename(columns={"Unnamed: 0":"Patient ID"},inplace=True) 
df
```

```python
##Setting new index
df.set_index("Patient ID")
```

```python
##Checking null values
df.isnull().sum()
```

```python
##Cleaning null values
df = df.dropna(how='any',axis=0) 
df
```

```python
##Convert float to int
float_col = df.select_dtypes(include=['float64']) 
float_col
```

```python
df['Kolku dolgo e dijabeticar'] = df['Kolku dolgo e dijabeticar'].astype('Int64')
df['Dali e na insulin ili na tabli?'] = df['Dali e na insulin ili na tabli?'].astype('Int64')
df['Dali ima dijabetesna retinopatija?'] = df['Dali ima dijabetesna retinopatija?'].astype('Int64')
df
```

### 3. Find the patient which currently has the highest glucose level. Sketch (on a plot of your choice) all of the patients glucose levels and describe the plot. Draw a conclusion on how the data behaves.

```python
#df.groupby(['Patient ID'], sort=False)['Glukoza (momentalna) vo krv'].max()
#df.sort_index(ascending=False)
patient1 = df.groupby('Patient ID')['Glukoza (momentalna) vo krv'].max().sort_values(ascending=False)##FALA BOGU

level = patient1.index[0]

glucose_level = patient1.iloc[0]

print(f"The patient {level} has the highest glucose leve,which is {glucose_level}.")
```

```python
patient2 = df.groupby('Patient ID')['Glukoza (momentalna) vo krv'].max().sort_values(ascending=False)

df2 = pd.DataFrame()

df2["Patient ID"] = patient2.index
df2["Glukoza (momentalna) vo krv"] = patient2.values

df2

#slika
fig = plt.figure(figsize=(20,20))

#Grafik
#sns.barplot(data=df2,x="Patient ID", y= "Glukoza (momentalna vo krv)")
plt.bar(patient2.index, patient2.values)

#prikaz
plt.show()
```

### 4. Find the average number of hemoglobin A1C levels of patients per length of disease. Which duration of the disease has the highest average glucose levels? Sketch the data on a plot of your choice to best describe the data. Write a sentence or two describing how the data behaves.

```python
result1 = (
    df.groupby(["Dali ima dijabetesna retinopatija?","Kolku dolgo e dijabeticar"], as_index=False)
      .agg(mean=('Kolku mu e tromesechen (hemoglobin A1C) sejker', 'mean'), count=('Kolku dolgo e dijabeticar', 'sum'))
)
result1
```

```python
patient4 = pd.DataFrame(result1,columns=["Dali ima dijabetesna retinopatija?","Kolku dolgo e dijabeticar","mean","count"])
patient4.plot(x="Dali ima dijabetesna retinopatija?",y=["mean","count"],kind="barh", figsize=(14,15))
plt.show()
```

### 5. Find the average duration of the disease for patients that are on insulin and patients that are on tablets. Compare the results. Which method that the patients use shows better results?


```python
result = (
    df.groupby(["Dali e na insulin ili na tabli?"], as_index=False)
      .agg(mean=('Kolku dolgo e dijabeticar', 'mean'), count=('Kolku dolgo e dijabeticar', 'count'))
)
result
#0  e na tabli
#1 e na insulin

```

---

```python
result.plot(x="Dali e na insulin ili na tabli?",y=["mean","count"],kind="bar")
plt.show()
print("Patients are using more tablets than insulin")
```

### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python
#mask = df["Kolku mu e tromesechen (hemoglobin A1C) sejker"] <= 5.7
#df1 = df["Kolku mu e tromesechen (hemoglobin A1C) sejker"]
#df2 = df[~mask]

#df1 patient that have normal A1C
#df2 patient that have elevated A1C 
df1,df2 = [x for _, x in df.groupby(df["Kolku mu e tromesechen (hemoglobin A1C) sejker"]>5.7)]
```

```python
series = df1.groupby("Patient ID")["Kolku dolgo e dijabeticar"].max().sort_values(ascending=False)

patient6 = series.index[0]

lenght_diabet = series.iloc[0]

print(f"The patient with number {patient6} is currently with the longest diabetic problem under 5.7glucose level with {lenght_diabet} years")
```

```python
series1 = df2.groupby("Patient ID")["Kolku dolgo e dijabeticar"].max().sort_values(ascending=False)

patient7 = series1.index[0]

lenght_diabet1 = series1.iloc[0]

print(f"The patient with number {patient7} is currently with the longest diabetic problem with 5.7glucose level above the normal with {lenght_diabet1} years")
```
